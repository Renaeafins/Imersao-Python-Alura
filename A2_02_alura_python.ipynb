```python
# Vamos agora aprender como criar um dataframe do zero!
# Chamaremos nosso Data Frame de 'salarios' e chamaremos a biblioteca do panda (pd) para isso
# Também usaremos a biblioteca 'numpy'
```

```python
import numpy as np

df_salarios = pd.DataFrame({
    'nome': ['Ana','Bruno','Carlos','Daniele','Val'],
    'salario': [4000,np.nan,5000,np.nan,100000]
})

# Tudo que for nulo preencheremos com a média do valor

# Calcula a média salarial e substitui os nulos pela média e arredonda os valores
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))

''' Pra fazer grandes comentários, dá pra usar 3 aspas antes e depois do comentaŕio, funciona em linhas separadas
'''
# Calcula a mediana e substitui os nulos (nan) por essa mediana
df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())
df_salarios
# df_salarios é nossa base - salario_media é a coluna que estamos criando nela - dentro dessa base, preencheremos o 'fillna' com a média (mean) dos salários - o round(2) arredonda as casas decimais
```

```python
df_temperaturas = pd.DataFrame({
    "Dia": ['Segunda','Terça','Quarta','Quinta','Sexta'],
    'Temperatura': [30,np.nan,np.nan,28,27] })
# Não precisamos importar o np novamente na linha acima, já importamos anteriormente

# Usaremos o Forward Fill (ffill) abaixo para preencher as próximas casaas com o valor anterior
df_temperaturas['preenchido_ffill'] = df_temperaturas['Temperatura'].ffill()
df_temperaturas
```

```python
df_temperaturas = pd.DataFrame({
    "Dia": ['Segunda','Terça','Quarta','Quinta','Sexta'],
    'Temperatura': [30,np.nan,np.nan,28,27] })
# Não precisamos importar o np novamente na linha acima, já importamos anteriormente

# Usaremos o Backwards Fill (bfill) abaixo para preencher casas anteriores com o valor seguinte/posterior, é o reverso do Forwards Fill (ffill)
df_temperaturas['preenchido_bfill'] = df_temperaturas['Temperatura'].bfill()
df_temperaturas
```

```python
df_cidades = pd.DataFrame({
    'nome': ['Ana','Bruno','Carlos','Daniele','Val'],
    'cidade': ['São Paulo',np.nan,'Curitiba',np.nan,'Belém'] })

df_cidades['cidade_preenchida'] = df_cidades['cidade'].fillna('Não informada')
df_cidades
```
